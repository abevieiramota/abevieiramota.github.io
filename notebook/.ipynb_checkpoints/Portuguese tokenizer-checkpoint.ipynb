```python
import nltk
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
import functools
import pandas as pd
```

```python
texto = """In sum I’m a geek interested in the social, in culture and politics. 
I have inherited a hands on mood from the BA in Graphic Design I hold. 
My passion for technology introduced me the idea of hacking."""
# fonte: http://cuducos.me/
```

# split + Counter

```python
split_counter = Counter(texto.split())
```

# nltk.word_tokenize + Counter

```python
nltk_counter = Counter(nltk.word_tokenize(texto, language='portuguese'))
```

# scikit-learn.feature_extraction.text.CountVectorizer

```python
cv = CountVectorizer(tokenizer=functools.partial(nltk.word_tokenize, language='portuguese'), lowercase=False)
texto_cv = cv.fit_transform([texto]).todense()
```

```python
sk_counter = dict(zip(cv.get_feature_names(), texto_cv.tolist()[0]))
```

# Comparando

```python
df = pd.DataFrame({'split_counter': split_counter, 'nltk_counter': nltk_counter, 'sk_counter': sk_counter})
```
